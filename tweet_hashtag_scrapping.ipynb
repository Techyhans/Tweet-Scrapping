##### Bulk Scrapping

**hashtag searched :** \
#workfromhome #wfh #pandemic #lockdown #covid #coronavirus #work #study #home #safe #stress #covid19 #C19 #struggle #reality #remotework #remoteclass #workingfromhome #studentlife #lifebalance #newnormal #newlifestyle #lifestyle #home #selfcare #stayhome

```python
import os
import pandas as pd

data = None
# data = pd.read_csv("hashtag.csv")
# len(data)
```

"workfromhome", "wfh", "pandemic", "lockdown", "covid", "coronavirus", "work", "study", 
           "home", "safe", "stress", "covid19", "C19", 'struggle', "reality", "remotework", "remoteclass", 
           "workingfromhome", "studentlife", "lifebalance", "newnormal", "newlifestyle", "lifestyle", 
           "home", "selfcare", "stayhome"

```python
hashtag = ["workfromhome", "wfh", "pandemic", "lockdown", "covid", "coronavirus", "work", "study", 
           "home", "safe", "stress", "covid19", "C19", 'struggle', "reality", "remotework", "remoteclass", 
           "workingfromhome", "studentlife", "lifebalance", "newnormal", "newlifestyle", "lifestyle", 
           "home", "selfcare", "stayhome"]
date_interval = ["2020-03-01", "2020-04-01", "2020-05-01", "2020-06-01", "2020-07-01", "2020-08-01", "2020-09-01",
                "2020-10-01", "2020-11-01", "2020-12-01", "2021-01-01", "2021-02-01", "2021-03-01", "2021-04-01", 
                "2021-05-01", "2021-06-01", "2021-06-26"]
for i in range(len(date_interval)-1):
    for hash in hashtag:
        os.system(
            f'snscrape --jsonl --progress --max-results 1000 --since {date_interval[i]} twitter-hashtag "{hash} until:{date_interval[i+1]}" > text-query-tweets.json')
        tweets_df = pd.read_json('text-query-tweets.json', lines=True)
        df = tweets_df[["id", "url", "date", "content",
                        "hashtags", "cashtags", "media", "lang"]]
        if data is None:
            data = df
        data = data.append(df)
        print(data)
```

```python
data.to_csv('hashtag.csv')
```
